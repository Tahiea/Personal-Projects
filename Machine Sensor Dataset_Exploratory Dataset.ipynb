As we are dealing with temperature and pressure for each feature here do they follow the natural law? That in a given volume the pressure and temperature are directly proportional? Let's see -

```python
f1.eval('k = Feature_1_Pressure/Feature_1_Temperature ', inplace=True)
f1
```

```python
sns.histplot(data = f1, x = 'k', bins=5)
```

#k column must have extreme values for the histogram to look like this.

```python
f1['k'].sort_values()
```

```python
f1.boxplot(column = 'k')
```

We will try to remove the outliers by calculating IQR

```python
Q1 = f1['k'].quantile(0.25)
Q3 = f1.k.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR

lower_limit, upper_limit
```

```python
#Records of k with outliers
outk = f1[(f1.k<lower_limit)|(f1.k>upper_limit)].sort_values(by = 'k')
outk
```

```python
sns.histplot(data = f1, x ='k', binrange = [0, 200], hue = 'Class', multiple="layer")
```

We can see that the pressure and the temperature is right skewed and the class types are more or less evenly distributed. So is the temperature and the pressure related? Maybe.
